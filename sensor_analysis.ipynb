```python
import numpy as np
import matplotlib.pyplot as plt

#generate time axis 

t = np.linspace(0,10,500)

# simulate clean signal

signal = np.sin (2*np.pi*1*t)

# add noise

noise = 0.3 * np.random.randn(len(t))
noisy_signal = signal + noise

# plot 

plt.plot (t, noisy_signal, label = "noisy signal")
plt.plot (t, signal, label = "clean signal")
plt.xlabel ("time")
plt.ylabel ("Amplitude")
plt.title ("simulated sensor") 
plt.show()


```

```python
# Moving Average Filter
window_size = 10
filtered_signal = np.convolve(noisy_signal, np.ones(window_size)/window_size, mode='same')

# Plot comparison
plt.plot(t, noisy_signal, label="Noisy Signal", alpha=0.5)
plt.plot(t, filtered_signal, label="Filtered Signal", linewidth=2)
plt.plot(t, signal, label="Clean Signal", linestyle='dashed')

plt.legend()
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Noise Reduction using Moving Average Filter")

plt.show()
```

```python
from scipy.signal import butter, filtfilt

# filter parameters

cutoff=2 # cutoof freqency Hz
fs=1/(t[1]-t[0])
order = 2 

# Design butterworh filter

b,a = butter(order,cutoff / (0.5 * fs),btype='low')
t = np.linspace(0, 10, 500)
# Apply filter 

butter_filtered = filtfilt(b,a, noisy_signal)

#plot comparision

plt.plot(t,noisy_signal, label="noisy signal", alpha=0.4)
plt.plot(t, butter_filtered, label="Butterworth Filtered", linewidth=2)
plt.plot(t, signal, label="Clean Signal", linestyle='dashed')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Low-Pass Butterworth Filter for Noise Reduction")

plt.show()
```
